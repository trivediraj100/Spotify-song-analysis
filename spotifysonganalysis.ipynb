```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Spotify_Youtube.csv")
```

```python
df.drop(columns = {"Unnamed: 0","Title" ,"Url_spotify", "Uri", "Url_youtube"}, inplace = True)
```

```python
df.nunique()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace = True)
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
def changefunc(likes):
    return int(likes)
df['Likes'] = df['Likes'].apply(changefunc)
```

```python
df['Views'] =df['Views'].astype('int')
df['Comments'] =df['Comments'].astype('int')
```

```python
df.dtypes
```

```python
# EDA
df.head(10)
```

```python
views = df.groupby('Artist')['Views'].sum().reset_index().sort_values('Views', ascending = False).head(10)
views
```

```python
plt.figure(figsize = (8,6))
plt.bar(views['Artist'] , views['Views'] , color = "green")
plt.grid()
plt.xlabel('artist')
plt.ylabel('views')
plt.title("artist vs no of songs")
plt.xticks(rotation = 70)
plt.show()
```

```python
Likes = df.groupby('Artist')['Likes'].sum().reset_index().sort_values('Likes' ,ascending = False).head(10)
plt.figure(figsize = (8,6))
plt.bar(Likes['Artist'], Likes['Likes'], color = "Green")
plt.xlabel('Artist')
plt.ylabel('Likes')
plt.xticks(rotation = 70)
plt.title("Likes Vs Artist")
plt.grid()
plt.show()
```
